```python
alpha = 0.0126843202700387
ref_temp = 20
data ['D1_CORR_1']= None
data ['D1_CORR_2']= None
data ['D1_CORR']= None
for index in data.index:
    data.loc [index, 'D1_CORR_1'] = data.loc [index, 'D1'] * np.exp (alpha * (ref_temp  - data.loc [index, 'Surface']))
    data.loc [index, 'D1_CORR_2'] = (0.9834*data.loc [index, 'D1_CORR_1'])-0.7718
    data.loc [index, 'D1_CORR'] = data.loc [index, ['D1_CORR_1','D1_CORR_2']].mean()
```
